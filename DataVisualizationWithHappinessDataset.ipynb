```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
style.use('seaborn')
plt.rcParams['font.weight'] = 'semibold'
plt.rcParams['font.family'] = 'georgia'
plt.rcParams['font.size'] = '16'
plt.rcParams['font.style'] = 'normal'
```

```python
df = pd.read_csv('datasets//world_happiness//world_happiness.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
pd.set_option('display.max_columns', 14)
pd.set_option('display.max_rows', 160)
```

```python
df = df.round(2)
```

```python
df.drop(columns=['Whisker.high', 'Whisker.low', 'Trust..Government.Corruption.', 'Dystopia.Residual'], inplace=True)
df
```

```python
#display the happiness scores per continent

group = df.groupby('Continent')['Happiness.Score'].agg('size').reset_index(name='happiness score')
group.to_csv('datasets//world_happiness//happiness_score_per_cont.csv')
group
```

```python
#visualize the distribution of happiness scores on a graoh to show how many countries have a happiness score between 5 and 6
bins = [2, 3, 4, 5, 6, 7, 8]
plt.hist(df['Happiness.Score'], bins=bins, edgecolor='#fafa66', color='#000000', alpha=0.8,
          label='happiness distribution', log=True)

plt.ylabel('frequency')
plt.xlabel('happiness score')
plt.legend(loc='upper left')
plt.yscale(value='linear')
plt.title('distribution of happiness scores')
plt.margins(0.2)
plt.show()
```

```python
#create a horizontal barchart with the combined top 10 and bottom 10 happy countries assuming the data in the table
#is not yet sorted out

top_10 = df[['Country', 'Happiness.Score']].sort_values('Happiness.Score', ascending=False).head(10)
bottom_10 = df[['Country', 'Happiness.Score']].sort_values('Happiness.Score', ascending=False).tail(10)

plt.barh(top_10['Country'], top_10['Happiness.Score'], color='#662211', label='top 10', edgecolor='black')
plt.barh(bottom_10['Country'], bottom_10['Happiness.Score'], color='#fa6699', label='bottom 10', edgecolor='black')

plt.legend()
plt.xlabel('happiness scores')
plt.title('top 10 vs bottom 10 happy countries')
plt.show()
```

```python
#create a barchart with the combined top 5 and bottom 5 countries sorted by generosity

top_5 = df[['Country',  'Happiness.Score', 'Generosity']].sort_values('Generosity', ascending=False).head(5)
bottom_5 = df[['Country', 'Happiness.Score', 'Generosity']].sort_values('Generosity', ascending=False).tail(5)

plt.bar(top_5['Country'], top_5['Happiness.Score'], color='brown', label='top 5', width=0.5,edgecolor='000000')
plt.bar(bottom_5['Country'], bottom_5['Happiness.Score'], color='black', label='bottom 5', width=0.5,edgecolor='#fafa66')

plt.ylabel('happiness score')
plt.title('Happiness score of top 5 vs bottom 5 generous countries')
plt.legend()
plt.xticks(rotation='vertical')
plt.show()
```

```python
#create a piechart showing the contributions of selected factors to happiness

slices = [df['Health..Life.Expectancy.'].sum(), df['Generosity'].sum(), df['Economy..GDP.per.Capita.'].sum(),
          df['Family'].sum(), df['Freedom'].sum()]
#colors = ['#9000bc', '#1277fa', '#9477fa','#002370','#1200bc','#fabcaa','#abba99','#fafa99','#83babc']
labels = ['health_expectancy', 'generosity', 'economic_gdp', 'family', 'freedom']
explode = [0,0,0,0.1,0]

plt.pie(slices, labels=labels, startangle=90, shadow=True, autopct='%1.1f%%', wedgeprops={'edgecolor': 'black'}, explode=explode)
plt.title('piechart showing the contributions of selected factors to happiness')
plt.show()
```

```python
top_10 = df[['Country', 'Economy..GDP.per.Capita.', 'Family', 'Health..Life.Expectancy.', 'Freedom', 'Generosity']].head(10)
```

```python
#create a stackplot showing the contributions of selected factors to happiness

top_10.plot(x = 'Country',kind='barh', stacked=True)
plt.legend(loc='upper right')
plt.xlabel('factor score')
plt.show()
```
