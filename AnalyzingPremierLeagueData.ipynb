```python
df.drop(columns='link_match', inplace=True)
```

```python
#fill the empty values in the passes_avg_A column with the avg values of the column

df['passes_avg_A'].fillna(df['passes_avg_A'].mean(), inplace=True)
df.round(2).head(100)
```

```python
#replace invalid/nan corners in the 'corners_avg_H with 0

df.loc[df['corners_avg_H'].isna(), 'corners_avg_H'] = 0
df.head(100)
```

```python
#drop nan rows

df.dropna(inplace=True)
df
```

```python
df.isna().sum()
```

```python
#create a new column for total goals scored

df['total_goals'] = df['goal_home_ft'] + df['goal_away_ft']
df
```

```python
#calculate the average goals per season

avg_goals_ps = round(df.pivot_table(index='season', values='total_goals', aggfunc='mean'), 2)
avg_goals_ps.to_csv('datasets//df_full_premierleague//avg_goals_ps.csv', sep=',')
avg_goals_ps
```

```python
#what are the biggest goal differences in games?

biggest_goal_diff_from_2010_to_2021 = abs(df['goal_home_ft'] - df['goal_away_ft']).sort_values(ascending=False).head(30)
biggest_goal_diff_from_2010_to_2021 = biggest_goal_diff_from_2010_to_2021.reset_index(name='goal_diff').drop(columns='index')
biggest_goal_diff_from_2010_to_2021.to_csv('datasets//df_full_premierleague//biggest_goal_diff_from_2010_to_2021.csv', sep=',')
biggest_goal_diff_from_2010_to_2021
```

```python
#what are the teams with the most away wins since 2010 to 2020

goal_diff = df['goal_away_ft'] > df['goal_home_ft']
wins = df.loc[goal_diff, ['away_team']].groupby('away_team').agg('size').sort_values(ascending=False).reset_index(name='most away wins')
wins.to_csv('datasets//df_full_premierleague//most_away_wins_from_2010_to_2021.csv', sep=',')
wins
```

```python
#what are the teams with the most goals scored at home

goals_scored = df.groupby('home_team')['goal_home_ft'].agg('sum').sort_values(ascending=False).reset_index()
goals_scored.to_csv('datasets//df_full_premierleague//most_home_goals_from_2010_to_2021.csv', sep=',')
goals_scored
```

```python
#what are the teams that received the least amount of goals while playing at home?

games_played = df.groupby('home_team').agg('size').reset_index(name='cnt')
goals_received = df.groupby('home_team')['goal_away_ft'].agg('sum').reset_index()
goal_ratio = goals_received['goal_away_ft'] / games_played['cnt']
goal_ratio.reset_index(name='avg_goal_received').join(df['home_team']).sort_values(by='avg_goal_received', ascending=True).to_csv('datasets//df_full_premierleague//least_amnt_goals_at_home.csv', sep=',')
goal_ratio.reset_index(name='avg_goal_received').join(df['home_team']).sort_values(by='avg_goal_received', ascending=True)
```
