```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
```

```python

df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df
```

```python
#select only companies located in seattle in alphabetical order

located = df['city'] == 'Seattle'
df.loc[located, ['company', 'city']].sort_values('company', ascending=[True])
```

```python
#provide companies with less than $80,000 in revenue

rev = df['revenue'] < 80000
new_rev = df.loc[rev, ['company', 'revenue']].sort_values(by='revenue', ascending=False)
new_rev
```

```python
#provide companies with more than $100,000 in revenue in dedcending order

rev = df['revenue'] > 100000
new_rev = df.loc[rev, ['company', 'revenue']].sort_values('revenue', ascending=[True])
new_rev.to_csv('datasets//fortune_1000_Conpanies//rvn_greater_100000.csv')
new_rev
```

```python
#select companies with the most and least revenue

rev = (df['revenue'] == df['revenue'].max()) | (df['revenue'] == df['revenue'].min())
news_rev = df.loc[rev, ['company', 'revenue'] ]
news_rev.to_csv('datasets//fortune_1000_Conpanies//most_least_rvn.csv')
news_rev
```

```python
#select companies with revenue between 80000 and 350000
rev = (df['revenue']> 80000) & (df['revenue']< 350000)
df.loc[rev, ['company', 'revenue']].sort_values('revenue', ascending=False)
```

```python
#how many companies have revenue greater than 40000

rev = df['revenue'] > 40000
print(df.loc[rev, ['company']].agg('size'), 'companies have revenue greater than $40000')
```

```python
#provide the top 10 companies with highest number of employees
num_emp = df[['company', 'num. of employees']].sort_values('num. of employees', ascending=[False]).head(10)
num_emp.to_csv('datasets//fortune_1000_Conpanies//top10_num_emps.csv')
num_emp
```

```python
#what is the relationship between the number of employees and the revenue of a company? 

#top 10 revenue generating companies
emp_rev_reltn = df[['company', 'revenue', 'num. of employees']].sort_values('revenue', ascending=[False]).head(10)
emp_rev_reltn.to_csv('datasets//fortune_1000_Conpanies//emp_rev_reltn.csv')
emp_rev_reltn
```

```python
#bottom 10 revenue generating companies
bottom_10 = df[['company', 'revenue']].sort_values('revenue', ascending=[False]).tail(10)
bottom_10.to_csv('datasets//fortune_1000_Conpanies//bott_10_revn.csv')
bottom_10
```

```python
#top 10 revenue generating companies
top_10 = df[['company', 'revenue']].sort_values('revenue', ascending=[False]).head(10)
top_10.to_csv('datasets//fortune_1000_Conpanies//top_10_revn.csv')
top_10
```

```python
#top 10 profit generating companies
top_10 = df[['company', 'profit']].sort_values('profit', ascending=[False]).head(10)
top_10.to_csv('datasets//fortune_1000_Conpanies//top_10_profit.csv')
top_10
```

```python
#bottom 10 profit generating companies
bottom_10 = df[['company', 'profit']].sort_values('profit', ascending=[False]).tail(10)
bottom_10.to_csv('datasets//fortune_1000_Conpanies//bott_10_profit.csv')
bottom_10
```

```python
df['Market Cap'] = df['Market Cap'].astype('float').round(0)
```

```python
#top 10 market capitals
top_10 = df[['company', 'Market Cap']].sort_values(by='Market Cap', ascending=False).head(10)
top_10.to_csv('datasets//fortune_1000_Conpanies//top_10_marketcap.csv')
top_10
```

```python
bottom_10 = df.groupby('company')['Market Cap'].agg('sum').reset_index().sort_values('Market Cap', ascending=True).head(10)
bottom_10.to_csv('datasets//fortune_1000_Conpanies//bott_10_marketcap.csv')
bottom_10
```
