```python
#drop records(rows) where the imdb_score column has missing values(NaN)

df.dropna(subset=['imdb_score'], inplace=True)
```

```python
df
```

```python
df.isna().sum()
```

```python
#for each tv show or movie which has nan value in the age certification column, replace it with 'no certification'

df['age_certification'].fillna('No certification', inplace=True)
```

```python
df
```

```python
df.isna().sum()
```

```python
#for each tv show or movie which has nan value in the seasons column, replace it with the most occured value in seasons

most_occured = df['seasons'].mode()[0]
most_occured
```

```python
df['seasons'].fillna(most_occured, inplace=True)
```

```python
df
```

```python
df.isna().sum()
```

```python
#count the number of movies or tvshows for each age certification

df.groupby('age_certification').size().reset_index(name='count')
```

```python
#count the number of movies and tv shows(seperately) produced in each release year

df.groupby(['release_year', 'type']).size().reset_index(name='count')
```

```python
#calculate the average runtime and imdb score of movies and tv shows for each release year

df.groupby(['release_year'])[['runtime', 'imdb_score']].mean().reset_index()
```

```python
#another way to do this

df.groupby(['release_year']).agg({'runtime': 'mean', 'imdb_score': 'mean'}).reset_index()
```

```python
#calculate the standard deviation of movies and tv shows ratings for each release year

df.groupby('release_year')['imdb_score'].std().reset_index()
```

```python
#calculate the standardized score for tmdb_popularity for each movie and tvshow within its respective genre
#and modify the original dataframe to include it

df['standardized_imdb'] = df.groupby('genres')['tmdb_popularity'].transform(lambda x: (x-x.mean())/x.std())
df
```

```python
#another way is to create the custom function and insert it into the transform method

def standardize(x):
    return (x-x.mean())/x.std()
    
df.groupby('genres')['tmdb_popularity'].apply(standardize).reset_index(name='standardized_imdb')
```

```python
#find the minimum, maximum, median and standard deviation release year for each movie or tvshow type

df.groupby('type')['release_year'].agg(['min', 'max', 'median', 'std'])
```

```python
#calculate the average and max imdb_score and tmdb_score for each season and release year combination

df.groupby(['release_year','seasons'])[['imdb_score', 'tmdb_score']].agg(['mean', 'max'])
```

```python
#calculate the average imdb_score and max tmdb_score for each season and release year combination

df.groupby(['seasons', 'release_year']).agg({'imdb_score': 'mean', 'tmdb_score': 'max'})
```

```python
df.groupby(['seasons', 'release_year'])['release_year'].agg('size').reset_index(name='count')
```

```python
#calculate the average length of titles(number of characters) for each type

def avg_len(titles):
    if isinstance(x, str):
        length = titles.apply(lambda x: len(x))
    return length.mean()

df.groupby(['genres'])['title'].transform(avg_len).reset_index()
```

```python
#find the count and average imdb_score for each age certification category

new_table = df.groupby(['age_certification'])['imdb_score'].agg(['size', 'mean'])
new_table.columns = ['count', 'avg_imdb']
new_table
```
